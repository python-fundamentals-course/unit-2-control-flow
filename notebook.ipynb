## 1. Making decisions with `if` / `elif` / `else`

Most business logic boils down to: **if this condition is true, do this; otherwise, do something else.** Python expresses that with `if`, `elif` ("else if"), and `else`.

Notice that the body of each branch is **indented** — Python uses indentation (not `{ }` or `end`) to mark which lines belong to which branch.


```python
transaction_amount = 750

if transaction_amount < 100:
    category = "small"
elif transaction_amount < 1000:
    category = "medium"
else:
    category = "large"

print(f"${transaction_amount} is a {category} transaction.")
```

## 2. Combining conditions with `and` / `or`

Real business rules often depend on more than one condition. You can combine comparisons with `and` (both must be true) and `or` (at least one must be true).


```python
income = 4500
credit_score = 690

if income >= 3000 and credit_score >= 650:
    decision = "approved"
elif income >= 3000 or credit_score >= 700:
    decision = "manual review"
else:
    decision = "declined"

print(f"Loan application: {decision}")
```

## 3. `for` loops — doing something for every item

A `for` loop runs its body once for each item in a sequence (like a list). This is the foundation of almost every data-processing task: you have many records, and you need to do *something* with each one.


```python
amounts = [25, 1200, 80, 450, 3000]

for amount in amounts:
    print(f"Processing transaction: ${amount}")
```

## 4. The three patterns: filter, transform, aggregate

Almost every data task is one of these three patterns (or a combination):

- **Filter** — keep only the items that match a condition
- **Transform** — turn each item into something new (e.g. reformat it)
- **Aggregate** — combine all the items into a single summary value (e.g. a total or count)

Let's apply all three to the same list of transactions.


```python
amounts = [25, 1200, 80, 450, 3000]

# Filter: keep only transactions over $100
large_transactions = []
for amount in amounts:
    if amount > 100:
        large_transactions.append(amount)

print("Large transactions:", large_transactions)
```

```python
# Transform: turn each amount into a formatted string
formatted = []
for amount in amounts:
    formatted.append(f"${amount:,.2f}")

print("Formatted:", formatted)
```

```python
# Aggregate: total and count
total = 0
count = 0
for amount in amounts:
    total = total + amount
    count = count + 1

print(f"Total: ${total}, Count: {count}, Average: ${total / count:.2f}")

# Python also has built-in shortcuts for common aggregates:
print(f"Using built-ins -> sum: {sum(amounts)}, count: {len(amounts)}, max: {max(amounts)}")
```

## 5. `while` loops — repeating until something changes

A `for` loop is for "do this for each item in a known collection". A `while` loop is for "keep doing this *until* a condition becomes false" — useful when you don't know in advance how many times you'll need to repeat.


```python
balance = 1000
monthly_withdrawal = 120
months = 0

while balance > 0:
    balance = balance - monthly_withdrawal
    months = months + 1

print(f"The balance runs out after {months} months (final balance: ${balance}).")
```

## 6. Decomposing a business problem

Before writing any code, it helps to break a problem into small steps written in plain English. For example:

> *"For each transaction, if the amount is over $1,000, flag it as 'review'. Count how many transactions were flagged, and calculate what percentage of all transactions that represents."*

Broken down:

1. Loop over each transaction (**for loop**)
2. Check if it's over the threshold (**if**)
3. Keep a running count of flagged transactions (**aggregate**)
4. After the loop, calculate the percentage

Here's the code that follows directly from that plan:


```python
transactions = [250, 1500, 80, 4200, 999, 1000, 50]
flagged_threshold = 1000

flagged_count = 0
for amount in transactions:
    if amount > flagged_threshold:
        flagged_count = flagged_count + 1
        print(f"  Flagged for review: ${amount}")

percentage = (flagged_count / len(transactions)) * 100
print(f"\n{flagged_count} of {len(transactions)} transactions flagged ({percentage:.1f}%)")
```
